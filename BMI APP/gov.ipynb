```python
import pandas as pd
import numpy as np
```

```python
CpIndex_df = pd.read_csv("All_India_Index.csv")
CpIndex_df
```

```python
# Identify and count missing values
missing_values = CpIndex_df.isnull().sum()
missing_values
```

#### Both the above methods by default consider all of the columns. You can also specify any subset of the DataFrame to detect duplicates.
#### By default, both keep the first observation in case of duplicates. We can specify "keep='last'" to instead keep the last observation.

```python
# a) Drop duplicate rows
CpIndex_df = CpIndex_df.drop_duplicates()
CpIndex_df
```

### b)Detect the outliers and remove the rows having outliers using IQR

```python
CpIndex_df.describe()
```

```python
# Calculate quartiles and IQR
Q1 = CpIndex_df['Egg'].quantile(0.25)
Q3 = CpIndex_df['Egg'].quantile(0.75)
IQR = Q3 - Q1
print(Q1)
print(Q3)
```

```python
# Define a condition to identify outliers (values outside the range [Q1 - 1.5*IQR, Q3 + 1.5*IQR])
outlier_condition = (CpIndex_df['Egg'] < Q1 - 1.5 * IQR) | (CpIndex_df['Egg'] > Q3 + 1.5 * IQR)

# Display the DataFrame with outliers
print("DataFrame with outliers:\n", CpIndex_df[outlier_condition])
```

```python
# Remove rows with outliers
CpIndex_df_no_outliers = CpIndex_df[~outlier_condition]

# Display the DataFrame after removing outliers
print("\nDataFrame after removing outliers:\n", CpIndex_df_no_outliers)
```

```python
# (c) Identify the most correlated positively correlated attributes and negatively correlated attributes
```

```python
CpIndex_df = CpIndex_df.drop(['Sector','Year','Month'], axis=1)
CpIndex_df
```

```python
# Calculate the correlation matrix
correlation_matrix = CpIndex_df.corr()
correlation_matrix
```

```python
# Extract the upper triangular part of the correlation matrix
upper_triangle = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))

# Most positively correlated attributes
positively_correlated = upper_triangle.abs().unstack().sort_values(ascending=False)
most_positively_correlated = positively_correlated[positively_correlated > 0].head(10)
print("Most Positively Correlated Attributes:")
print(most_positively_correlated)
```

```python
# Most negatively correlated attributes
negatively_correlated = upper_triangle.abs().unstack().sort_values(ascending=True)
most_negatively_correlated = negatively_correlated[negatively_correlated > 0].head(10)
print("\nMost Negatively Correlated Attributes:")
print(most_negatively_correlated)
```
