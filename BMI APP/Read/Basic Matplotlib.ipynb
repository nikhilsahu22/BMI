```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax2.plot([1.5, 3.5, -2, 1.6])   # plot ax2 using x-axis as index array 0..N-1
ax3.plot([1.5, 2, 4, 1.6])
ax1.plot(np.random.randn(50).cumsum(),linestyle='--',marker='o',color='y')

# When you don't specify a specific subplot (Axes object) using the ax parameter, 
# Matplotlib creates a new plot in the current Axes, which is the last created Axes or 
# the default one.
plt.plot(np.random.randn(100).cumsum(),linestyle='--',marker='x',color='r',label='A' )

plt.legend()
plt.show()

# supported values for linestyle are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

Markers are symbols that can be placed at data points. You can use the marker parameter in 
the plt.plot() function to specify a marker style.

```python
# Sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Plot with markers
plt.plot(x, y, marker='o', linestyle='-', color='k', label='Data with Markers')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot with Markers')
plt.legend()

# Show the plot
plt.show()
```

Annotations can be used to add text or arrows to the plot. You can use the plt.annotate() function for this purpose

```python
# Sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Plot the data
plt.plot(x, y, label='Data')

# Annotate a specific point              #specify loc. of text label
plt.annotate('Maximum', xy=(np.pi/2, 1), xytext=(np.pi/2, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot with Annotation')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

### Vertical and Horizontal Lines

#### plt.axvline() --> add vertical and
#### plt.axhline() --> horizontal lines to the plot

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
# Plot the data
plt.plot(x, y, label='Data')

# Add a vertical line
plt.axvline(x=np.pi/2, color='r', linestyle='--', label='Vertical Line')
plt.axhline(y=0, color='r', linestyle='-', label='Horizantal Line')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot with Vertical Line')
plt.legend()

# Show the plot
plt.show()
```

### Histograms

#### The bins parameter controls or tells you the number of bins that your data will be divided into. You can specify it as an integer or as a list of bin edges .personal favorite is "Bayesian Blocks" (bins="blocks")

######    align: Alignment of bars ('left', 'mid', 'right').    
######    orientation: Orientation of the histogram ('vertical', 'horizontal').

######    label: Label or sequence of labels.After that must--->  plt.legend()   
######    stacked: If True, multiple data are stacked on top of each other.

##### Density Plot with Matplotlib

###### alpha controls the transparency of the bars.

######  plot provides a visual comparison between the histogram of the random data and the probability density function of a normal distribution fitted to that data


```python
from scipy.stats import norm
# Generate random data
data = np.random.randn(1000)
# Create a density plot
plt.hist(data, bins=30, density=True, edgecolor='black', color='skyblue', alpha=0.7)


# Plot the fitted normal distribution----->

xmin, xmax = plt.xlim()     # define the limits of the x-axis
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))

# pdf=probability density function------->
plt.plot(x, p, 'k', linewidth=2)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Plot')

# Show the plot
plt.show()
```

## Basic Scatter Plot with Matplotlib

#### x and y: Data for the x and y axes.
#### s: Marker size./ want differentsize pass as variable and create list s' of that s=s'

```python
# Generate random data
x = np.random.randn(100)
y = 2 * x + np.random.randn(100)

# Create a scatter plot
plt.scatter(x, y, color='blue', alpha=0.7)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

# Show the plot
plt.show()
```

### Scatter Plot with Categorical Colors:

```python
# Generate random data
x = np.random.randn(100)
y = 2 * x + np.random.randn(100)

categories = np.where(x > 0, 'r', 'g')

# Create a scatter plot with categorical colors
plt.scatter(x, y, c=categories, cmap='coolwarm', alpha=0.7)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Categorical Colors')
# Show the plot
plt.show()
```
