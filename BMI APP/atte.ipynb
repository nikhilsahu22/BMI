Q4.Consider two excel files having attendance of a workshop’s participants for two days. Each file has three
fields ‘Name’, ‘Time of joining’, duration (in minutes) where names are unique within a file. Note that
duration may take one of three values (30, 40, 50) only. Import the data into two dataframes and do the
following:
a. Perform merging of the two dataframes to find the names of students who had attended the workshop on
both days.
b. Find names of all students who have attended workshop on either of the days.
c. Merge two data frames row-wise and find the total number of records in the data frame.
d. Merge two data frames and use two columns names and duration as multi-row indexes. Generate
descriptive statistics for this multi-index.


```python
import pandas as pd
import numpy as np

# DataFrame for Day 1
df1_data = {
    'Name': ['Rohit', 'Aman', 'Yash', 'David', 'Shri'],
    'Time of Joining': np.random.choice(pd.date_range('2023-11-20 08:00:00', '2023-11-20 10:00:00')),
    'Duration': np.random.choice([30, 40, 50], 5)
}
df1_data = pd.DataFrame(df1_data)
df1_data 
```

```python
# DataFrame for Day 2
df2_data = {
    'Name': ['Rohit', 'Aman', 'Yash', 'David', 'Sid'],
    'Time of Joining': np.random.choice(pd.date_range('2023-11-21 08:00:00', '2023-11-21 10:00:00')),
    'Duration': np.random.choice([30, 40, 50], 5)
}
df2_data = pd.DataFrame(df2_data)
df2_data
```

```python
# a. Perform merging of the two dataframes to find the names of students who had attended the workshop on both days.

both_days_attendees = pd.merge(df1_data , df2_data , on='Name', how='inner') # inner:inter of keys from both frames
print("Names of Students who Attended the Workshop on Both Days:\n", both_days_attendees['Name'])
```

```python
# b. Find names of all students who have attended workshop on either of the days.

either_day_attendees = pd.merge(df1_data , df2_data , on='Name', how='outer') # outer: use union of keys from both frames
print("Names of Students who Attended the Workshop on Either of the Days:\n", either_day_attendees['Name'])
```

```python
# c. Merge two data frames row-wise and find the total number of records in the data frame.

merged_row_wise = pd.concat([df1_data , df2_data ], ignore_index=True)
merged_row_wise
```

```python
print("Total Number of Records after Merging Row-Wise:", len(merged_row_wise))
```

```python
# d. Merge two data frames and use two columns names and duration as multi-row indexes. Generate descriptive statistics for this multi-index.

merged_multi_index = pd.merge(df1_data , df2_data , on=['Name', 'Duration'], how='outer')
statistics_multi_index = merged_multi_index.groupby(['Name', 'Duration']).describe()
print("\nDescriptive Statistics for Multi-Index (Names and Duration):\n", statistics_multi_index)
```
